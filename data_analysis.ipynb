```python
import matplotlib.pyplot as plt
import pandas as pd

host_path = "training_data/2020_05_04/host/"

oracle = pd.read_csv(host_path+'db_oracle_11g.csv')
container = pd.read_csv(host_path+'dcos_container.csv')
docker = pd.read_csv(host_path+'dcos_docker.csv')
redis = pd.read_csv(host_path+'mw_redis.csv')
linux = pd.read_csv(host_path+'os_linux.csv')

host_data = pd.concat([oracle,container,docker,redis,linux])
```

```python
host_data = host_data.sort_values(by=['timestamp'])
```

```python
print(len(pd.unique(linux.name)))
_linux = linux[linux.name==pd.unique(linux.name)[12]]

plt.hist(_linux.value, bins = 20)
plt.show()
```

```python
def dora_s_test(x):
    return all(x==max(x))

useless_kpis = []

for kpi, group in host_data.groupby('name'):
    if dora_s_test(group.value):
        useless_kpis.append(kpi)
        

for ukpi in useless_kpis:
    print('KPI: ' + ukpi + '\'s values are all ' + str(int(pd.unique(host_data[host_data['name']==ukpi].value))))
```
